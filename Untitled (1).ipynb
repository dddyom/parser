```python
artist_works = pd.read_csv('product.csv')
```

```python
artist_works['dimension_cm'] = artist_works[['length', 'width', 'height']].apply(
    lambda x: ' x '.join(x.dropna().astype(str)),
    axis=1
)
artist_works['dimension_cm'] = artist_works['dimension_cm'] +' см'
```

```python
artist_works = artist_works[~artist_works['size_currency'].isin(['2012', '2018'])]
```

```python
artist_works.loc[(artist_works['price_1_currency'] == 'euro'), 'price_1_currency'] = '€'
artist_works.loc[(artist_works['price_2_currency'] == 'dollar'), 'price_2_currency'] = '$'
artist_works.loc[(artist_works['price_3_currency'] == 'ruble'), 'price_3_currency'] = '₽'
```

```python
artist_works.genre.value_counts()
```

```python
artist_works.loc[(artist_works['genre'] == 'zhivopis'), 'genre'] = 'Живопись'
artist_works.loc[(artist_works['genre'] == 'printi'), 'genre'] = 'Принт'
artist_works.loc[(artist_works['genre'] == 'graphika'), 'genre'] = 'Графика'
artist_works.loc[(artist_works['genre'] == 'object'), 'genre'] = 'Объект'
artist_works.loc[(artist_works['genre'] == 'peyzazh'), 'genre'] = 'Пейзаж'
artist_works.loc[(artist_works['genre'] == 'photographii'), 'genre'] = 'Фотография'
artist_works.loc[(artist_works['genre'] == 'portret'), 'genre'] = 'Портрет'
artist_works.loc[(artist_works['genre'] == 'collazh'), 'genre'] = 'Коллаж'
artist_works.loc[(artist_works['genre'] == 'video'), 'genre'] = 'Видео'
artist_works.loc[(artist_works['genre'] == 'abstraktsiya'), 'genre'] = 'Абстракция'
```
